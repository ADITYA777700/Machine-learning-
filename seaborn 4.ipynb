```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
```

```python
var=sns.load_dataset("titanic")
var
```

```python
sns.pairplot(var,vars=["fare","age"],hue="sex",hue_order=["female","male"],kind="kde",diag_kind="kde")
plt.title("pairplot")
plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
sns.stripplot(x="class",y="fare",data=var,hue="sex",palette="rocket_r",linewidth=1,edgecolor="g",jitter=0.5,size=5
              , marker="^",alpha=1)
plt.title("striplot")
plt.show()
```

```python
sns.stripplot(x=var["fare"])
plt.show()
```

```python
sns.set(style="whitegrid")
sns.boxplot(x="class",y="fare",data=var,hue="sex",color="b",order=["Third","first","second"],linewidth=3,palette="rocket_r"
           )
plt.title("boxplot")
plt.show()
```

```python
sns.set(style="whitegrid")
sns.boxplot(y=var["fare"])
plt.show()
```

```python
sns.catplot(x="class",y="fare",data=var,hue="sex",palette="BuPu",kind="bar")
plt.figure(figsize=(12,5))
plt.title("catplot")
plt.show()
```
